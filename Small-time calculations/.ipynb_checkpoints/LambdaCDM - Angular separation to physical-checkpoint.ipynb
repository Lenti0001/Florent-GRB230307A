```python
from astropy.cosmology import FlatLambdaCDM
import numpy as np
import matplotlib.pyplot as plt
```

```python
cosmo = FlatLambdaCDM(H0=67.7, Om0=0.31)
```

```python
print(cosmo) #Our cosmological model (Planck 2015).
```

```python
#Cross-check - we want to see that the physical distance per angular separation increases then decreases again, 
#as is expected of a cosmological model
#This is indeed the case - neat!

z = np.linspace(0, 10, 1000)
plt.plot(z, cosmo.kpc_proper_per_arcmin(z).value)
plt.grid()
plt.xlabel("z [N/A]")
plt.ylabel("Physical distance per angular separation [kpc/']")
```

```python
### Calculations for the first potential host galaxy (G1). Separation from GRB230307A is 30.2 arcsec and z is 0.0646
Separation_kpc_G1 = (cosmo.kpc_proper_per_arcmin(0.06).value)*(30.2/60)
print("Separation from G1 (if z is similar): " + str(Separation_kpc_G1) + " kpc")
```

```python
### Calculations for the background galaxy. Separation from GRB230307A here is 0.2 arcsec and z is 3.86
Separation_kpc_G1 = (cosmo.kpc_proper_per_arcmin(3.87).value)*(0.3/60)
print("Separation from background galaxy (if z is similar): " + str(Separation_kpc_G1) + " kpc")
```
