```python
from astropy.io import fits
import matplotlib.pyplot as plt
import numpy as np
```

```python
#fits_image_filename = 'coadded_1dspec.fits'
fits_image_filename = '1dspec_total.fits'

hdul = fits.open(fits_image_filename)
```

```python
hdul[11].data["TRACE_SPAT"] ### Spat-pos x ~= 1146 is the transient object in the 1dspec_total.fits file case!
```

```python
obj = hdul[11] ## ... Therefore we select the 11th index as the object!
```

```python
obj.data['OPT_FLAM']
obj.data['OPT_FLAM_IVAR']
```

```python
wavelengths =  obj.data['OPT_WAVE']
fluxes = obj.data['OPT_FLAM']
flux_errors = obj.data['OPT_FLAM_SIG']
flux_var = obj.data['OPT_FLAM_IVAR']
```

```python
plt.xlim(4000, 9000)
plt.plot(wavelengths, fluxes, color = 'black', label="fluxes")
plt.plot(wavelengths, flux_var, color = 'orange', label="fluxvar")
plt.xlabel('Wavelength (Angstroms)')
plt.ylabel('Flux (1e-17 erg/s/cm**2/AA)')
plt.legend()
plt.show()
```

```python
plt.ylim(-10, 10)
plt.xlim(5500, 6250)
plt.plot(wavelengths[300:1500], fluxes[300:1500], color = 'black', label="fluxes")
plt.plot(wavelengths[300:1500], flux_var[300:1500], color = 'orange', label="fluxvar")
plt.plot(wavelengths[300:1500], flux_errors[300:1500], color="grey", label="flux_errors")
plt.xlabel('Wavelength (Angstroms)')
plt.ylabel('Flux (1e-17 erg/s/cm**2/AA)')
plt.legend()
plt.show()
```
